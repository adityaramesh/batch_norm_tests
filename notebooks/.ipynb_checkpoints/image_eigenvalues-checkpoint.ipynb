```python
def make_hist(data, bins=50):
    hist, edges = np.histogram(np.log(data), bins=bins)
    fig = plt.figure(width=600, height=600)
    fig.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
        fill_color="#036564", line_color="#033649")
    return fig
```

```python
# First plot the histograms of the singular values.

datasets = ["cifar_10", "stl_10", "svhn"]

for ds in datasets:
    file = h5py.File("data/svd_results/{}.hdf5".format(ds), 'r')
    s = np.array(file["singular values"])
    fig = make_hist(s, bins=50)
    plt.show(fig)
```

```python
# First idea: plot the normalized cumulative explained variance. The plot shows
# the percentage of the variance explained by the first $n$ components.

def make_explained_var_plot(data):
    cum = np.cumsum(data**2)
    cum = cum / cum[-1]
    fig = plt.figure(width=600, height=600)
    fig.line(np.arange(len(cum)), cum)
    return fig

for ds in datasets:
    file = h5py.File("data/svd_results/{}.hdf5".format(ds), 'r')
    s = np.array(file["singular values"])
    fig = make_explained_var_plot(s)
    plt.show(fig)
```

```python
# Same as above but with axes reversed, so that we see the number of components
# needed to retain x% of the total variance.

def make_explained_var_plot(data):
    cum = np.cumsum(data**2)
    cum = cum / cum[-1]
    fig = plt.figure(width=600, height=600)
    fig.line(cum, np.arange(len(cum)))
    return fig

for ds in datasets:
    file = h5py.File("data/svd_results/{}.hdf5".format(ds), 'r')
    s = np.array(file["singular values"])
    fig = make_explained_var_plot(s)
    plt.show(fig)
```

```python
# Plot the polynomial fitted to the empirical CDF.

from scipy.optimize import curve_fit

# Things that didn't work:
# * Exponential: f(x) = a - b * exp(-c * x). To enforce that f(0) = 0, we set b = a.
# * Hyperbolic tangent.
#
# Instead we try rational functions. TODO: find a way to regularize the coefficients
# of the linear terms in both the numerator and the denominator. They tend to blow up
# and cause bad behavior near 0.
def func(x, p0, p1, p2, p3, p4, q0, q1, q2, q3, q4):
    return np.polyval([p0, p1, p2, p3, p4, 1], x) / np.polyval([q0, q1, q2, q3, q4, 0], x)

def make_cdf(data):
    x = sorted(data)
    y = np.arange(len(data)) / len(data)
    popt, _ = curve_fit(func, x, y)
    yhat = func(x, *popt)
    print(popt)
    
    fig = plt.figure(width=600, height=600)
    fig.line(x, y, color="black")
    fig.line(x, yhat, color="blue")
    return fig

for ds in datasets:
    file = h5py.File("data/svd_results/{}.hdf5".format(ds), 'r')
    s = np.array(file["singular values"])
    fig = make_cdf(s)
    plt.show(fig)
```

```python
"""
Ok, so we've successfully fit a rational function to the CDF. Now let's derive
an analytic expression for the curvature of the rational function using sympy.
"""

from sympy import *
from IPython.display import display

init_printing(use_latex=True)
```

```python
from fractions import Fraction
from sympy.functions import Abs

x = Symbol("x")
p0, p1, p2, p3, p4 = symbols("p_0 p_1 p_2 p_3 p_4")
q0, q1, q2, q3, q4 = symbols("q_0 q_1 q_2 q_3 q_4")

y = (p0 * x**5 + p1 * x**4 + p2 * x**3 + p3 * x**2 + p4 * x + 1) / \
    (q0 * x**5 + q1 * x**4 + q2 * x**3 + q3 * x**2 + q4 * x)
yp = diff(y, x)
ypp = diff(y, x, 2)
yppp = diff(y, x, 3)
kappa = ypp / (1 + yp**2)**Fraction(3/2)
r = Abs((1 + yp**2)**Fraction(3/2) / ypp)

display(yp)
```

```python
# Define functions for convenient/fast evaluation of the expressions.

feval = lambdify((x, p0, p1, p2, p3, p4, q0, q1, q2, q3, q4), y)
der_1 = lambdify((x, p0, p1, p2, p3, p4, q0, q1, q2, q3, q4), yp)
der_2 = lambdify((x, p0, p1, p2, p3, p4, q0, q1, q2, q3, q4), ypp)
curvature = lambdify((x, p0, p1, p2, p4, p3, q0, q1, q2, q3, q4), kappa)
radius = lambdify((x, p0, p1, p2, p3, p4, q0, q1, q2, q3, q4), r)
```

```python
from scipy.optimize import brentq

def get_params(data):
    x_ = sorted(data)
    y_ = np.arange(len(data)) / len(data)
    popt, _ = curve_fit(func, x_, y_)
    return popt

def make_cdf_2(data):
    x = sorted(data)
    y = np.arange(len(data)) / len(data)
    fig = plt.figure(width=600, height=600)
    fig.line(x, y, color="black")
    return fig

def make_hist_with_line(data):
    popt = get_params(data)
    
    def score(x_):
        # The intuition here is based on the mean value theorem. We want to
        # find a point in the CDF at which the derivative matches the average
        # slope over the entire CDF.
        return der_1(x_, *popt) - 1 / len(data)
    
    # Use this for root-finding.
    
    # The derivative at 0 isn't defined, so don't use it. The optimizer will reject
    # this value.
    res = brentq(score, a=30, b=len(data), maxiter=1000)
    idx = np.argmax(np.flipud(data) > res)
    
    var = np.flipud(data**2)
    ratio = var[var < var[idx]].sum() / var.sum()
    print("Cutoff: {} (index {}).".format(res, idx))
    print("Variance explained before cutoff: {}%.".format(100 * ratio))
    print("Fraction before cutoff: {}%.".format(idx / len(data)))
    print("Singular value at cutoff: {}.".format(np.flipud(data)[idx]))
    print("Mean, median: {}, {}.".format(np.mean(data), np.median(data)))
    
    #x_ = np.arange(len(data))
    #y_ = np.array([score(z) for z in x_])
    
    #fig = plt.figure(width=600, height=600)
    #fig.line(x_, y_)
    #return fig
    
    #hist = make_hist(data, bins=100)
    #hist.line([data[idx], data[idx]], [0, 3000])
    #return hist
    
    cdf = make_cdf_2(data)
    cdf.line([res, res], [0, 1])
    return cdf

for i, ds in enumerate(datasets):
    file = h5py.File("data/svd_results/{}.hdf5".format(ds), 'r')
    s = np.array(file["singular values"])
    fig = make_hist_with_line(s)
    plt.show(fig)
```
